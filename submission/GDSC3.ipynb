```python
# from sklearn.decomposition import PCA
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.svm import SVC
# from sklearn.svm import SVR
# from sklearn.ensemble import ExtraTreesClassifier
# from xgboost import XGBClassifier
# from sklearn.ensemble import AdaBoostClassifier
# from sklearn.linear_model import LinearRegression
# from sklearn.linear_model import LogisticRegression
# from sklearn.linear_model import Ridge
# from sklearn.cluster import KMeans
# from sklearn.impute import KNNImputer
# from sklearn.model_selection import cross_val_score
# from sklearn.model_selection import RandomizedSearchCV
# from scipy import stats
# from sklearn.preprocessing import PowerTransformer
# from sklearn.preprocessing import QuantileTransformer

# importing all the required libraries and modules

import pandas as pd
import numpy as np
import warnings
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier

warnings.filterwarnings('ignore')

#importing the training data set
train_df = pd.read_csv(r"C:\Users\lenovo\gdsc-ml-nsut-recruitment-comp-2024\train.csv")
train_df = train_df.dropna(subset=["TARGET"]) #dropping rows where TARGET column has NaN values
train_df
```

```python
#importing the test data set
test_df = pd.read_csv(r"C:\Users\lenovo\gdsc-ml-nsut-recruitment-comp-2024\test (1).csv")
test_df
```

```python
x_train = train_df.drop(columns=['ID','TARGET'],axis=1) #assigning required features
y_train = train_df.TARGET
x_test = test_df.drop('ID',axis=1)

imputer = SimpleImputer(strategy='median') #imputing the missing values with median strategy
x_trained = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)
y_trained = y_train

x_trained = np.log(x_trained + 1)  # adding 1 to handle zero values
x_test = np.log(x_test + 1)
```

```python
scaler = StandardScaler(with_std=False) #initializing the scaler without standard deviation (zero mean only)
x_train2, x_test2, y_train2, y_test2 = train_test_split(x_trained,y_trained,test_size=0.2) #splitting the dataset into train and test 
x_train_scaled = scaler.fit_transform(x_train2) #scaling the data
x_test_scaled = scaler.transform(x_test2)
x_test_ans = scaler.transform(x_test)

#Hyper Parameter Tuning
model_temp = GridSearchCV(RandomForestClassifier(),{  #using GridSearchCV to find the best parameters
    'n_estimators':[2,3,4],
    'max_depth':[5,10],
    'min_samples_split':range(1, 6)
} , cv=5, return_train_score=False)

model_temp.fit(x_train_scaled,y_train2)
model_temp.best_score_, model_temp.best_params_
```

```python
output = 0  
max_attempts = 50
attempt = 0

#running a loop to capture the most accurate model
while output <= 0.939 and attempt < max_attempts:  # loop until the score exceeds 0.939 or max attempts are reached
    
    x_train2, x_test2, y_train2, y_test2 = train_test_split(x_trained,y_trained,test_size=0.2) #splitting the dataset into train and test 
    x_train_scaled = scaler.fit_transform(x_train2) #scaling the data every iteration
    x_test_scaled = scaler.transform(x_test2)
    x_test_ans = scaler.transform(x_test)

    #using RandomForestClassifier with the best performing parameters found using Grid Search CV and hit-and-trial method to train the model
    model = RandomForestClassifier(n_estimators=4, criterion='log_loss', max_depth=5, min_samples_split=3) 
    bag_model = BaggingClassifier(estimator=model, n_estimators=8) #stacking model using BaggingClassifier 
    bag_model.fit(x_train_scaled,y_train2) 
    
    output = bag_model.score(x_test_scaled,y_test2) 

    print(f"Attempt {attempt + 1}: Current score: {output}")  #printing each iteration
    attempt += 1

if output > 0.939:
    print("HES DONE IT!")
else:
    print("Try Again")
```

```python
sub=pd.read_csv(r"C:\Users\lenovo\gdsc-ml-nsut-recruitment-comp-2024\sample.csv") #replacing the TARGET column with the values of the model predicted
sub['TARGET'] = bag_model.predict(x_test_ans)
sub.to_csv("submission_demo24.csv",index=False)
sub
```
